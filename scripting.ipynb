```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Churn_Modelling.csv")
print(df.columns)
df=df.drop(columns=["RowNumber","CustomerId","Surname"])
print(df.describe)
# we have dropped the coloumns which aren't important
```

```python

numeric_columns=["CreditScore","Age","Tenure","Balance","NumOfProducts","EstimatedSalary"]
boolean_columns=["HasCrCard","IsActiveMember","Exited"]
alpha_coulmns=["Geography","Gender"]
#here we have divided the coloumns into their various parts
```

```python
print(df["Exited"].value_counts())
print(df["Exited"].value_counts(normalize=True))

sns.countplot(data=df, x="Exited")


# here we have done the churn rate analysis
```

```python
for i in boolean_columns:
    sns.countplot(data=df,x=i)
    plt.show()
    print(df[i].value_counts())

#here we have worked on counting the values of how many of the people have the credit cards whether they are active members and are a customer with the bank or not 
```

```python
for i in alpha_coulmns:
    sns.histplot(data=df,x=i)
    plt.show()
#this is the description of diffrent customers of where they are from and what's their gender
```

```python
for i in numeric_columns:
    sns.histplot(data=df,x=i,kde=True)
    plt.show()
#this is to show how the values are distributed along in the numeric values 
```

```python
#now here we starrt with bivariate analysis and analyse various parts of the dataset 

sns.countplot(data=df,x="Geography",hue="Exited")
plt.show()
print(df.groupby("Geography")["Exited"].mean())
print("Germany had a higher churn rate ")
#here the average churn rate of each country is determined and anaylysed


```

```python
sns.countplot(data=df ,x="Gender", hue="Exited")
plt.show()
print(df.groupby("Gender")["Exited"].mean())
print("Females had a higher churn rate")
```

Co-relation Heatmap 

```python
numeric_df = df.select_dtypes(include="number")

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
```
